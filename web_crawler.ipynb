```python
def savedata():
    url = input("please input url:")#get the web information about player's statistic    
    urlstr = str(url)
    res = requests.get(urlstr)
    soup = BeautifulSoup(res.text, "lxml")

    data = []
    k=["Age","Tm","Lg","G","PA","AB","R","H","2B","3B","HR","RBI","SB","CS","BB","SO","BA","OBP","SLG","OPS","OPS+","TB","GDP","HBP","SH","SF","IBB","Pos","Awards"]
    
    table = soup.find('tbody')
    rows = table.find_all('tr', 'full')
    for row in rows:
        cols = row.find_all('td')
        cols = [ele.text.strip() for ele in cols]
        data.append([ele for ele in cols]) 
        
    frame = pd.DataFrame(data,columns=k)
    
    uc = urlparse(url)
    name_split = uc.path.split("/")[3]
    name = name_split.replace(".shtml","")
    filename = "{}.xlsx".format(name)
    
    print(frame)
    print(name)
    writer = pd.ExcelWriter(filename, engine='xlsxwriter')
    frame.to_excel(writer, sheet_name='Sheet1')
    writer.save()
    return frame
    return filename
```

```python
url = "https://world.taobao.com/search/json.htm?navigator=all&_ksTS=1517495786342_300&spm=a21bp.7806943.20151106.1&search_type=0&_input_charset=utf-8&json=on&q=%E8%91%AB%E8%8A%A6%E4%BF%9D%E6%B8%A9%E6%9D%AF%E5%A4%9A%E6%A0%B7%E5%B1%8B&cna=IFGhEWrB0l8CATHac%2FB4wtYx&callback=__jsonp_cb&abtest=_AB-LR517-LR854-LR895-PR517-PR854-PR895&nid=&type=&uniqpid="
res = requests.get(url)
```

```python
print(res.text)
```

```python
import json
import re
#把json格式去除一些不必要的東西，主要是開頭與結尾的格式
m = re.search("if\(window.__jsonp_cb\)\{__jsonp_cb\((.*?)\)\}", res.text)
jd = json.loads(m.group(1))
```

```python
jd
```

```python
#因為格式太複雜，所以重新將它存成json黨，在用瀏覽器開啟，它會自動將它排版
with open("taobau.json", "w") as f:
    f.write(json.dumps(jd))
```
